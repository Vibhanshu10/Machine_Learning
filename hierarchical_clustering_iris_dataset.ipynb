```python
X=df["sepal_length"]
Y=df["petal_length"]
```

```python
data=list(zip(X.values,Y.values))
data
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
model = AgglomerativeClustering(n_clusters=2)
```

```python
model.fit(data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,c=model.labels_)
plt.show
```
